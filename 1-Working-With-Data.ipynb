# Working with Data
When doing anything related to machine learning, working with data is a fundamental concept that needs to be understood. Cleansing data so that it is in a consistent format that contains the important pieces that you're interested in is paramount to successful implementations.

Data is typically provided by a database or a file that can represent data in a tabular format. One such file type is CSV(Comma Separated Files).

## Representation of Data
The best way to represent data as we know it is tables and rows within those tables. Think MS Excel.

Each column in a table represents a differnt piece of data. Each row in the table consists of individial records that relate all the values in the different columns to a single relationship.

In the example below, we're looking at some data from the Iris flowser dataset. Each row represents a different variation of speicies of the Iris flower, and each column represents a property of that record.

The data may also seem strange. Its the Iris Flowser dataset -  a common dataset used to learn data concepts. This dataset describes different species of Iris data flowers by their properties.

![https://www.math.umd.edu/~petersd/666/html/iris_with_labels.jpg](https://www.math.umd.edu/~petersd/666/html/iris_with_labels.jpg)

Notice that each column has a data type associated to it. Double is a decimal value. It is important to understand the data type of each cell in a table to be able to work with it. Imagine we thought the values were strings, we wouldn't be able to do calculations with them. This is also important for the code we write. Programming lanuages need to know what data types working with to run the correct operations on them.

## Data in Files

Here's an example of some data from a comma seperated file. Notice that it's the same as a table, it has columns that specify the meaning of the data, and rows that represent individual records.

```
Id,SepalLengthCm,SepalWidthCm,PetalLengthCm,PetalWidthCm,Species
1,5.1,3.5,1.4,0.2,Iris-setosa
2,4.9,3.0,1.4,0.2,Iris-setosa
3,4.7,3.2,1.3,0.2,Iris-setosa
```

Notice that the first row indicates the headers for the values that correspond to that repective item in the sequence. This CSV file is comma serperated. It is also important to note that the data values use "." for decimals rather than commas. Things could be very confusing if they did.

The code below simply reads the file as it is. It does not concern itself with the purpose of the file. You could try read in any file. 


````python
with open("datasets/Iris.csv") as csv_file:
    for row in csv_file:
        print(row)
````

The csv package in Python helps in working with CSV files specifically. The code below is very similar to the code for reading a file. However, it allows for reading CSV files as lists that we can work with in Python rather than just rows of data that we need to process ourselves. 

````python
import csv

csv_file = open("datasets/Iris.csv", newline="")
reader = csv.reader(csv_file, delimiter=",")
for row in reader:
    print(row)

````

Now we have our data rows as arrays. We can work with this in an easy way. Let's try print the row number with the species of flower per row.

````python
csv_file.seek(0)
next(reader)  # Skip the first row
for row in reader:
    print(row[0], "\t", row[5])  # Format the output for easier reading
        
````

We can also load the data into a 'data frame'. This isn't a native Python data type, but rather comes from a library called Pandas.

Data frames are commonly used for data processing in languages such as Python and R, as they will have many functions designed for working with large sets of data, more than lists, which are a lower-level data type.

Here we read the csv file into a data frame and print out the first 5 rows in it.

````python
import pandas as pd

iris_df = pd.read_csv("datasets/Iris.csv")
iris_df.head()
````

We can split data frames into smaller data frames, selecting some of the rows to go into the new data frame, some of the columns or both. Note that this does not change the old data frame, just creates a new one that is a subset of the old.

Here we take just the Id and Species columns into a new data frame.

````python
split_df = iris_df[['Id', 'Species']]
split_df.head()
````

And here we create a new data frame with only those rows where the Sepal Length is 5.0 or higher

````python
split_df = iris_df[iris_df["SepalLengthCm"]>= 5.0] 
split_df.head()
````

We can modify data within a data frame, updating specific values

````python
# iris_df[iris_df['Id']==1]['Species'] = 'Test'
iris_df.loc[0, 'Species'] = 'Test'
iris_df.loc[3, 'Species'] = 'Test'
iris_df.head()
````

Or values that match specific conditions

````python
iris_df.loc[iris_df['Species'] == 'Test', 'Species'] = "Iris-setosa"
iris_df.head()
````
